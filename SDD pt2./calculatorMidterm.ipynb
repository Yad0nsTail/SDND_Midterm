```python
import unittest
import script
import math

class Testscript(unittest.TestCase):
    
    #Test 3 : Cosine

    def test_cos(self):
        self.assertAlmostEqual(script.cos(0), 1.0, places=2)
        self.assertAlmostEqual(script.cos(math.pi), -1.0, places=2)
        self.assertAlmostEqual(script.cos(math.pi/2), 0.0, places=2)
        self.assertAlmostEqual(script.cos(math.pi/4), 0.707, places=2)
    
if __name__ == '__main__':
    unittest.main(argv=['first-arg-is-ignored'], exit=False)
```

```python
import unittest
import script

class Testscript(unittest.TestCase):
    
    #Test 5: Tangent

    def test_tan(self):
        self.assertAlmostEqual(script.tan(0), 0.0, places=2)
        self.assertAlmostEqual(script.tan(math.pi), 0.0, places=2)
        self.assertAlmostEqual(script.tan(math.pi * 2), 0.0, places=2)
        self.assertAlmostEqual(script.tan(math.pi * -1 ), 0.0, places=2)
        
if __name__ == '__main__':
    unittest.main(argv=['first-arg-is-ignored'], exit=False)
```
